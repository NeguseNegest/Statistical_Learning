(c) För varje trick 1−4 skapa en ny kolumn med namnet ’make i’ för i = 1, 2, 3, 4 så att värdet av ’make i’ i en given rad är 1 om skateboardåkaren landade trick i och 0 annars

```python

# Create the new columns initialized with NaN values
make_columns = ["make 1", "make 2", "make 3", "make 4"]
for make_column in make_columns:
    df[make_column] = np.nan

# Populate the 'make i' columns based on the 'trick i' columns
for i, make_column in enumerate(make_columns, start=1):
    trick_column = f'trick {i}'
    df[make_column] = df[trick_column].apply(lambda x: 1 if x > 0 else 0)

df
```

(d) För varje skateboardåkare skatta sannolikheten att ett trick får ett betyg
som är större än 0.6 givet att skateboardåkaren landar tricket. Vad är sannolikheten att skateboardåkaren inte lyckas landa ett visst trick? Vad observerar
du? Relatera dina observationer till era observationer i del (b).


Vi har enligt bayes sats att $P(A|B)=\frac{P(B|A)P(A}{P(B)}$

Om vi ansätter $L=$Lyckat trick
och $S$ som ett poäng större än 0.6.

får vi $P(S|L)=\frac{P(L|S)P(S)}{P(L)}$

Men $P(L|S)P(S)=1*P(S)$ så vi får utrycket

$P(S|L)=\frac{P(S)}{P(L)}$


```python

contestants = df["id"]
for contender in contestants:
    P_L = sum(df[df['id'] == str(contender)].loc[:, 'make 1':'make 4'].values[0].tolist())
    d = df[df['id'] == str(contender)].loc[:, 'trick 1':'trick 4'].values[0].tolist()
    dummy = [x for x in d if x > 0.6]
    P_S = len(dummy)

    if P_L == 0:
        prob = 0
    else:
        prob = P_S/P_L

    print("Probability for %s to get a higher score than 0.6 given he landed a trick is %f" % (contender, prob))
```
